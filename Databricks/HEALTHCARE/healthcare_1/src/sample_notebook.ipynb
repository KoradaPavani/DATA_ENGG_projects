This default notebook is executed using a Lakeflow job as defined in resources/sample_job.job.yml.

```python
spark.sql("CREATE CATALOG IF NOT EXISTS healthcare_new")
spark.sql("USE CATALOG healthcare_new")
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS healthcare_db_sql_1")
spark.sql("USE healthcare_db_sql_1")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS Patients (
  patient_id INT,
  first_name STRING,
  last_name STRING,
  dob DATE,
  gender STRING,
  phone STRING,
  email STRING,
  address STRING
)
USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS Doctors (
  doctor_id INT,
  doctor_name STRING,
  phone STRING,
  email STRING,
  department_id INT
)
USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS Appointments (
  appointment_id INT,
  patient_id INT,
  doctor_id INT,
  appointment_date DATE,
  appointment_time TIMESTAMP,
  status STRING
)
USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS Diagnoses (
  diagnosis_id INT,
  appointment_id INT,
  diagnosis_details STRING,
  diagnosis_date DATE
)
USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS Prescriptions (
  prescription_id INT,
  appointment_id INT,
  prescription_date DATE
)
USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS Medications (
  medication_id INT,
  medication_name STRING,
  dosage STRING,
  instructions STRING
)
USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS Prescription_Medications (
  prescription_id INT,
  medication_id INT,
  quantity INT
)
USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS Lab_Tests (
  lab_test_id INT,
  appointment_id INT,
  test_name STRING,
  test_date DATE,
  result STRING
)
USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS Billing (
  bill_id INT,
  appointment_id INT,
  amount INT,
  payment_status STRING,
  payment_date DATE
)
USING DELTA
""")
```

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS Insurance (
  insurance_id INT,
  patient_id INT,
  provider_name STRING,
  policy_number STRING,
  coverage_details STRING
)
USING DELTA
""")
```
